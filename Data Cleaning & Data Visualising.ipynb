# Importing Necessary Libraries

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

# Load all datasets 

```python
content = pd.read_csv("Content.csv")
reactions = pd.read_csv("Reactions.csv")
ReactionTypes=pd.read_csv('ReactionTypes.csv')
```

# Data Cleaning

```python
# getting relevent columns of content table 
```

```python
content.head()
```

```python
content_category_typecontent=content[['Content ID' , 'User ID' , 'Type' , 'Category']]
content_category_typecontent.head()
```

```python
# getting relevent columns from reactions table 
```

```python
reactions.head()
```

```python
content_typereaction=reactions[['Content ID' , 'Type' , 'Datetime']]
content_typereaction.head()
```

```python
# getting relevent columns from reaction types table
```

```python
ReactionTypes.head()
```

```python
reaction_score=ReactionTypes[['Type' , 'Sentiment' , 'Score']]
reaction_score.head()
```

```python
# merge all datasets into one dataset
```

```python
content_type_score=content_typereaction.merge(reaction_score , on='Type' )
content_type_score.head()
```

```python
content_type_score
```

```python
final_data=content_category_typecontent.merge(content_type_score , on='Content ID' , suffixes=['_content' , '_reaction'])
final_data.head()
```

```python
final_data.isnull().sum()
```

```python
final_data = final_data.dropna()
final_data.info()
```

```python
final_data['Datetime'] = pd.to_datetime(final_data['Datetime'])
final_data.info()
```

```python
final_data.head()
```

```python
final_data['Category'].value_counts()
```

- Searching the categories we find erroneous values like "dogs" is normally dogs and "food" and Food that must be replaced by food 
- So here we should clean our category columns from this erroneous values because it's gonna affect the final insights
- We have just 16 categories but there is repeated ones

```python
#address the reapted column (especially for categorical varibale)!!
final_data['Category'] = final_data['Category'].replace('"','',regex = True)
final_data['Category'] = final_data['Category'].str.lower()
```

```python
final_data['Category'].unique()
```

```python
final_data.to_csv('final_data.csv', index=False)
```

```python
final_data.head()
```

```python
final_data.groupby('Category').agg('sum', 'Score').sort_values('Score', ascending = False).head(5)
```

# So the top 5 categories are: 
- animals
- science
- healthy eating
- techonology 
- food

# Data Visualising

```python
final_data.info()
```

```python
# Calculate the sum of scores for each category
category_popularity = final_data.groupby('Category')['Score'].sum().sort_values(ascending=False)

# Select the top 5 categories
top_5_categories = category_popularity.head(5)

# Reverse the color map to make the bar with the highest score darkest
colors = plt.cm.Oranges(np.linspace(1, 0.2, len(top_5_categories)))

# Plot the top 5 categories with custom colors
plt.figure(figsize=(10, 6))
plt.bar(top_5_categories.index, top_5_categories, color=colors)
plt.title('Top 5 Content Categories by Popularity')
plt.xlabel('Category')
plt.ylabel('Popularity (Sum of Score)')
plt.show()
```
